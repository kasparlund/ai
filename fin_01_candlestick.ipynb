```python
# export
from math import pi

import pandas as pd

from bokeh.io import output_notebook, show
from bokeh.plotting import figure #, output_file
from bokeh.sampledata.stocks import MSFT
output_notebook() #inline plotting
```

```python
# export
def plotCandlestick(df, ohlc_columns=[DATE,OPEN,HIGH,LOW,CLOSE], title=None, width=None, height=None ):
    inc = df[CLOSE] > df[OPEN]
    dec = df[OPEN]  > df[CLOSE]
    w   = 12*60*60*1000 # half day in ms

    TOOLS = "pan,wheel_zoom,box_zoom,reset,save"
    
    if height is None:
        p = figure(x_axis_type="datetime", tools=TOOLS, plot_width=width, title = title)
    else: 
        p = figure(x_axis_type="datetime", tools=TOOLS, plot_width=width, plot_height=height, title = title)
    p.xaxis.major_label_orientation = pi/4
    p.grid.grid_line_alpha=0.3

    p.segment(df[DATE], df[HIGH], df[DATE], df[LOW], color="black")
    p.vbar(df.loc[inc,DATE], w, df.loc[inc,OPEN], df.loc[inc,CLOSE], fill_color="#D5E1DD", line_color="black")
    p.vbar(df.loc[dec,DATE], w, df.loc[dec,OPEN], df.loc[dec,CLOSE], fill_color="#F2583E", line_color="black")

    show(p)  # open a browser
```

```python
[DATE,OPEN,HIGH,LOW,CLOSE] = ["date","open","high","low", "close"]
df = pd.DataFrame(MSFT)[:50]
df["date"] = pd.to_datetime(df["date"])
plotCandlestick(df, title="MSFT Candlestick", width=1000, height=200)
```
