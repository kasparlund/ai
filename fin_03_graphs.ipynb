```python
# export
from math import pi

import pandas as pd
import numpy as np

from bokeh.io import output_notebook, show
from bokeh.layouts import gridplot
#from bokeh.plotting import figure #, output_file
from bokeh.plotting import figure, output_file, show
from bokeh.sampledata.stocks import MSFT
output_notebook() #inline plotting
```

```python
# export 

TOOLS = "pan,wheel_zoom,box_zoom,reset,save,box_select"
def plotSignalVSPrediction( x, graph_data, legendTitle="Legends" ):
    p1 = figure(title="input vs predictions", tools=TOOLS)
    
    for (signal, title, color, shape) in graph_data:
        if shape=="circle":
            p1.circle(x, signal, legend_label=title, color=color )
        else:    
            p1.line(x,   signal, legend_label=title, color=color )
        #p1.line(x, prediction, legend_label="2*sin(x)", color="orange")
        #p1.circle(x, 3*y, legend_label="3*sin(x)", color="green")
        
        #p1.legend.title = legendTitle

    show(gridplot([p1], ncols=1, plot_width=1000, plot_height=400))  # open a browser
```

```python
# export 

def make_plot(title, hist, edges, x):
    p = figure(title=title, tools='', background_fill_color="#fafafa")
    p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:],
           fill_color="navy", line_color="white", alpha=0.5)

    p.y_range.start = 0
    #p.legend.location = "center_right"
    #p.legend.background_fill_color = "#fefefe"
    p.xaxis.axis_label = 'x'
    p.yaxis.axis_label = 'Pr(x)'
    p.grid.grid_line_color="white"
    return p
```

```python
x = np.linspace(0, 4*np.pi, 100)
y = np.sin(x)
  
open_reference_data  = (y,"open reference","blue","line")
open_prediction_data = (y,"open prediction","orange","circle")
plotSignalVSPrediction(x,[open_reference_data,open_prediction_data])
```
